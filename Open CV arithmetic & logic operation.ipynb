```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
img1 = cv2.imread('3D-Matplotlib.png')
img2 = cv2.imread('mainsvmimage.png')
```

```python
add = cv2.add(img1,img2)
plt.imshow(add)
plt.show()
```

```python
add = img1+img2
plt.imshow(add)
plt.show()
```

```python
weighted = cv2.addWeighted(img1,0.3,img2,0.7,3)
plt.imshow(weighted)
plt.show()
```
