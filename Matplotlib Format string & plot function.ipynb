```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7,8]
y=[23,54,63,34,12,71,43,51]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color="green",linewidth=6,linestyle='dashed')
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color="green",linewidth=6,linestyle='dashed')
```

```python
plt.plot(x,y,'b+--')
```

```python
plt.plot(x,y,'gD--')
```

```python
# Same effect as 'gD--' format string
plt.plot(x,y,color='green',marker='D',linestyle='--')
```

```python
plt.plot(x,y,color='green',marker='D',linestyle='--',markersize='20')
```

```python
plt.plot(x,y,color='green',marker='D',linestyle='--',markersize='20',alpha=0.5)
```
