```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.jointplot(data = tips , x = 'total_bill' , y = 'tip' , kind = 'reg')
```

```python
sns.jointplot(data = tips , x = 'total_bill' , y = 'tip' , kind = 'hex')
```

```python
sns.jointplot(data = iris , x = 'petal_length' ,y = 'petal_width' , kind = 'reg' , color = 'green')
```

```python
sns.jointplot(data = iris , x = 'petal_length' ,y = 'petal_width' , kind = 'hex' , color = 'green')
```

```python
sns.jointplot(data = tips , x = 'total_bill' , y = 'tip' , kind = 'kde')
```
