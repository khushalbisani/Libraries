```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

# WebCam Image Capture

```python
def main():
    cap = cv2.VideoCapture(0)
    
    if cap.isOpened():
        ret , frame = cap.read()
        print(ret)
        print(frame)
        
    else:
        ret = False
        
        
    img = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    
    plt.imshow(img)
    plt.title("Khushal")
    plt.xticks([])
    plt.yticks([])
    plt.show()
    cap.release()
    
    
```

```python
if __name__ == "__main__":
    main()
```

# Live video feed capture and processing

```python

def main():
    counter = 0
    window = "Khushal"
    cv2.namedWindow(window)
    cap = cv2.VideoCapture(0)
    
    if cap.isOpened():
        ret , frame = cap.read()
        
    else:
        ret = False
        
    while True:
        ret , frame = cap.read()
        frame = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY)
        
        cv2.imshow(window , frame)
        
        counter+=1
        if cv2.waitKey(1) == 27:
            break
    
    cap.release()        
    cv2.destroyAllWindows()
    print(counter)
    
```

```python
if __name__ == "__main__":
    main()
```

# Webcam video recording file

```python
def main():
    windowname = "live webcam video feed capturing"
    cv2.namedWindow(windowname)
    cap = cv2.VideoCapture(0)
    
    filename = 'C:\\Users\\khush\\Documents\\OpenCV\\output\\vid.avi' ####
    codec = cv2.VideoWriter_fourcc('W' , 'M' ,'V' ,'2') ####
    framerate = 30 ####
    resolution = (640 , 480) ####
    
    VideoFileOutput = cv2.VideoWriter(filename , codec , framerate , resolution) ####
    
    if cap.isOpened():
        ret , frame = cap.read()
    else:
        ret = False
        
    while ret:
        ret , frame = cap.read()
        frame = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
        VideoFileOutput.write(frame) ####
        cv2.imshow(windowname , frame)
        
        if cv2.waitKey(1) == 27:
            break
            
    cv2.destroyAllWindows()
    VideoFileOutput.release() ####
    cap.release()
        

```

```python
if __name__ == "__main__":
    main()
```

# Play video using video player

```python
def main():
    namedwindow = 'video player'
    cv2.namedWindow(namedwindow)
    filename = "C:\\Users\\khush\\Documents\\OpenCV\\output\\vid.avi"
    cap = cv2.VideoCapture(filename)
    
    while(cap.isOpened()):
        
        ret , frame = cap.read()
        
        print(ret)
        
        if ret:
            cv2.imshow(namedwindow , frame)
            if cv2.waitKey(33) == 27:
                break
                
        else:
            break
            
    cv2.destroyAllWindows()
    cap.release()
                
```

```python
if __name__ == "__main__":
    main()
```
