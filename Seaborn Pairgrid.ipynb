```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
iris = sns.load_dataset("iris")
```

```python
n = sns.PairGrid(data = iris)
n = n.map(plt.scatter)
```

```python
n = sns.PairGrid(data = iris)
n = n.map_diag(plt.hist)
```

```python
n = sns.PairGrid(data = iris)
n = n.map_diag(plt.hist)
n = n.map_offdiag(plt.scatter)
```

```python
n = sns.PairGrid(data = iris , hue = 'species')
n = n.map_diag(plt.hist)
n = n.map_offdiag(plt.scatter)
```

```python
n = sns.PairGrid(data = iris , hue = 'species' , palette = 'rocket')
n = n.map_diag(plt.hist)
n = n.map_offdiag(plt.scatter)
n = n.add_legend()
```

```python
n = sns.PairGrid(data = iris , hue = 'species' , palette = 'rocket')
n = n.map_diag(plt.hist , histtype='step' ,linewidth = 5)
n = n.map_offdiag(plt.scatter)
n = n.add_legend()
```

```python
x = sns.PairGrid(data = iris , vars = ["petal_length" , "petal_width"] , hue = 'species')
x = x.map_offdiag(plt.scatter)
x = x.map_diag(plt.hist)
x = x.add_legend()
```

```python
x = sns.PairGrid(data = iris , vars = ["petal_length" , "petal_width"] , hue = 'species')
x = x.map_offdiag(plt.scatter , edgecolor = 'blue')
x = x.map_diag(plt.hist , edgecolor = 'black')
x = x.add_legend()
```

```python
x = sns.PairGrid(data = iris , x_vars = ['petal_length','petal_width'] , y_vars = ['sepal_length','sepal_width'])
x = x.map(plt.scatter)
```

```python
n = sns.PairGrid(data = iris , hue = 'species' , palette = 'rocket')
n = n.map_diag(plt.hist)
n = n.map_upper(plt.scatter)
n = n.map_lower(sns.kdeplot)
n = n.add_legend()
```

```python
n = sns.PairGrid(data = iris , hue = 'species' , hue_kws = {'marker' : ['d','s','+']})
n = n.map(plt.scatter,s=20)
n = n.add_legend()
```
