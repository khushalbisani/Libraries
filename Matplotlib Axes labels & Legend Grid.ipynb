```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6]
a=[23,54,67,78,43,47]
b=[34,56,62,80,89,12]
c=[12,23,45,56,67,78]
```

```python
plt.xlabel("time")
plt.ylabel("utilization")
plt.title("Career")

plt.plot(x,a)
plt.plot(x,b)
plt.plot(x,c)
```

```python
plt.xlabel("time")
plt.ylabel("utilization")
plt.title("Career")

plt.plot(x,a,label='a')
plt.plot(x,b,label='b')
plt.plot(x,c,label='c')
plt.legend(loc='best')
```

```python
plt.plot(x,a,label='a')
plt.plot(x,b,label='b')
plt.plot(x,c,label='c')
plt.legend(loc="upper right",shadow=True,fontsize='large')
```
