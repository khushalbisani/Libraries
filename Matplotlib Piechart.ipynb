```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
quantity=[23,45,21,5,67]
food=['Apple','Mango','Grapes','Pineapple','Guava']
```

```python
plt.pie(quantity,labels=food)
```

```python
plt.pie(quantity,labels=food)
plt.show
```

```python
plt.axis("equal")
plt.pie(quantity,labels=food)
plt.show
```

```python
plt.axis("equal")
plt.pie(quantity,labels=food,radius=1.5)
plt.show
```

```python
plt.axis("equal")
plt.pie(quantity,labels=food,radius=1.5,autopct="%f%%")
plt.show
```

```python
plt.axis("equal")
plt.pie(quantity,labels=food,radius=1.5,autopct="%0.2f%%")
plt.show
```

```python
plt.axis("equal")
plt.pie(quantity,labels=food,radius=1.5,explode=[0,0.15,0,0.1,0])
plt.show
```

```python
plt.axis("equal")
plt.pie(quantity,labels=food,radius=1.5,autopct="%f%%",explode=[.2,0,0,0,0],counterclock=True,startangle=45)
plt.show
```

```python
plt.axis("equal")
plt.pie(quantity,labels=food,explode=[.2,0,0,0,0],startangle=45)
plt.show
```

```python
plt.axis("equal")
plt.pie(quantity,labels=food,explode=[.2,0,0,0,0],startangle=45)
plt.savefig("piechart.png",bbox_inches='tight',pad_inches=1,transparent=True)
```

```python
plt.axis("equal")
plt.pie(quantity,labels=food,explode=[.2,0,0,0,0],startangle=45)
plt.savefig("piechart.pdf",bbox_inches='tight',pad_inches=1)
```
