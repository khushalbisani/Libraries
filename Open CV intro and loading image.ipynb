```python
img = cv2.imread('IMG-20181101-WA0022.jpg',cv2.IMREAD_COLOR)

cv2.line(img , (0,0) , (150,150) , (255,0,0) , 8 )

cv2.rectangle(img , (20,40),(250,300),(0,255,0),5)

cv2.circle(img , (30,60),7,(0,0,255),6)

ar=np.array([[25,30],[5,5],[120,154],[178,181]] , np.int64)
cv2.polylines(img,[ar],True,(0,0,255),4)

font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img , 'Mayank' , (100,300) , font , 5, (123,234,0) , 5)

cv2.imshow('Friend',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
