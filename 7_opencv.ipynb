```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
%matplotlib inline
```

```python
def main():
    subpath = "C:\\Users\\khush\\Documents\\OpenCV\\misc\\"
    imgpath =  subpath + "4.1.03.tiff"
    img = cv2.imread(imgpath)
    img = cv2.cvtColor(img  , cv2.COLOR_BGR2RGB)
    
    p = 0.02
    
    row , column , channel = img.shape
    output = np.zeros(img.shape , np.uint8)
    
    for i in range(row):
        for j in range(column):
            r = random.random()
            if r < p/2:
                output[i][j] = [0,0,0]
            elif r < p:
                output[i][j] = [255,255,255]
            else:
                output[i][j] = img[i][j]
                
    plt.subplot(1,2,1)
    plt.imshow(img)
    plt.title("Orignal img")
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(1,2,2)
    plt.imshow(output)
    plt.title("noise of salt and pepper image")
    plt.xticks([])
    plt.yticks([])
```

```python
if __name__ == "__main__":
    main()
```

# Low Pass Filter

```python
def main():
    subpath = "C:\\Users\\khush\\Documents\\OpenCV\\misc\\"
    imgpath =  subpath + "4.1.03.tiff"
    img = cv2.imread(imgpath)
    img = cv2.cvtColor(img  , cv2.COLOR_BGR2RGB)
    

    box = cv2.boxFilter(img , -1 , (23,23))          ###########
    blur = cv2.blur(img , (23,23) )                  ###########
    gaussian = cv2.GaussianBlur(img , (23,23) , 0)   ###########
    
    output = [img , box , blur , gaussian]
    title = ["Img" , "Box" , "Blur" , "Gaussian"]
        
    for i in range(4):
        plt.subplot(1,4,i+1)
        plt.imshow(output[i])
        plt.title(title[i])
        plt.xticks([])
        plt.yticks([])
```

```python
if __name__ == "__main__":
    main()
```

# Median blur filter

efficient for removing salt and pepper noise from the image

```python
def main():
    subpath = "C:\\Users\\khush\\Documents\\OpenCV\\misc\\"
    imgpath =  subpath + "4.1.03.tiff"
    img = cv2.imread(imgpath)
    img = cv2.cvtColor(img  , cv2.COLOR_BGR2RGB)
    
    p = 0.02
    
    row , column , channel = img.shape
    noised = np.zeros(img.shape , np.uint8)
    
    for i in range(row):
        for j in range(column):
            r = random.random()
            if r < p/2:
                noised[i][j] = [0,0,0]
            elif r < p:
                noised[i][j] = [255,255,255]
            else:
                noised[i][j] = img[i][j]
                
    denoised = cv2.medianBlur(noised , 5) # IT IS VERY NICE WAY TO REMOVE SALT AND PEPPER NOISE OF IMAGE
                
    plt.subplot(1,3,1)
    plt.imshow(img)
    plt.title("Orignal img")
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(1,3,2)
    plt.imshow(noised)
    plt.title("noise of salt and pepper image")
    plt.xticks([])
    plt.yticks([])
    
    plt.subplot(1,3,3)
    plt.imshow(denoised)
    plt.title("DEnoise of salt and pepper image")
    plt.xticks([])
    plt.yticks([])
```

```python
if __name__ == "__main__":
    main()
```
