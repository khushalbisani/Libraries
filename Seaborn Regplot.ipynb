```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.regplot(data = tips , x='tip' , y='total_bill')
```

```python
mean=[2,3]
variance = [[1.1 , 0.3],[2.2 , 3]]
x_value , y_value = np.random.multivariate_normal(mean , variance , 100).T
```

```python
sns.regplot(x=x_value ,y=y_value)
```

```python
series_x_value = pd.Series(x_value , name = 'x-way')
series_y_value = pd.Series(y_value , name = 'y-way')

sns.regplot(x=series_x_value , y=series_y_value , marker = 'D' , color = 'grey')
```

```python
sns.regplot(x=series_x_value , y=series_y_value ,\
           scatter_kws = {'color' : 'red'},\
           line_kws = {'color' : 'blue' , 'linewidth' : 6})
```

```python
sns.regplot(x=series_x_value , y=series_y_value, marker = 'D' ,\
           scatter_kws = {'color' : 'grey' },\
           line_kws = {'color' : 'black' , 'linewidth' : 3})
```

```python
sns.regplot(x=series_x_value , y=series_y_value, ci =60 )
```

```python
sns.regplot(x=series_x_value , y=series_y_value ,x_estimator = np.mean)
```
