```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
n = sns.FacetGrid(data = tips , row = 'smoker' , col = 'time')
n = n.map(plt.hist , 'total_bill')
```

```python
n = sns.FacetGrid(data = tips , row = 'smoker' , col = 'time')
n = n.map(plt.hist , 'tip', color = 'blue' , bins = 15)
```

```python
x = sns.FacetGrid(data = tips , row = 'day' , col = 'time')
x = x.map(plt.hist , 'tip' , color = 'green')
```

```python
y = sns.FacetGrid(data = tips , row = 'sex' , col = 'smoker')
y = y.map(plt.scatter , 'total_bill' , 'tip')
```

```python
y = sns.FacetGrid(data = tips , row = 'smoker' , col = 'sex')
y = y.map(plt.scatter , 'total_bill' , 'tip')
```

```python
y = sns.FacetGrid(data = tips , row = 'sex' , col = 'smoker')
y = y.map(sns.regplot , 'total_bill' , 'tip')
```

```python
x = sns.FacetGrid(data = tips , col = 'sex',hue = 'smoker')
x = x.map(plt.scatter , 'total_bill' , 'tip')
x = x.add_legend()
```

```python
x = sns.FacetGrid(data = tips , col = 'day')
x = x.map(sns.boxplot , 'sex' ,'total_bill' )
```

```python
x = sns.FacetGrid(data = tips , col = 'day',size=15 , aspect =0.5)
x = x.map(sns.boxplot , 'sex' ,'total_bill' )
```
