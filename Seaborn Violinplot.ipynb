```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.violinplot(data = tips)
```

```python
sns.violinplot(data = tips , x = 'day' , y = 'total_bill')
```

```python
sns.violinplot(data = tips , x = 'day' , y = 'total_bill' , hue = 'sex')
```

```python
sns.violinplot(data = tips , x = 'sex' , y = 'tip')
```

```python
sns.violinplot(data = tips , x = 'sex' , y = 'total_bill')
```

```python
sns.violinplot(data = tips , x = 'day' , y = 'total_bill' , inner = 'quartile')
```

```python
sns.violinplot(data = tips , x = 'day' , y = 'total_bill' ,hue = 'sex' , inner = 'quartile')
```

```python
sns.violinplot(data = tips , x = 'day' , y = 'total_bill' ,hue = 'sex' , inner = 'quartile' , split = True)
```

```python
sns.violinplot(data = tips , x = 'day' , y = 'total_bill' ,hue = 'sex' , inner = 'quartile' , split = True , scale = 'count')
```

```python
sns.violinplot(data = tips , x = 'day' , y = 'total_bill' ,hue = 'sex' , inner = 'quartile' , split = False , scale = 'count')
```

```python
sns.violinplot(data = tips , x = 'day' , y = 'total_bill' ,hue = 'sex' , inner = 'stick', scale = 'count')

# each black line show a line of a data point
```

```python
sns.violinplot(data = tips , x = 'day' , y = 'total_bill' ,hue = 'sex' , inner = 'stick', scale = 'count' , bw = 0.3)
```

```python
sns.violinplot(data = tips , x = 'day' , y = 'total_bill' , inner = 'quartile' , scale = 'count')
sns.swarmplot(data = tips , x = 'day' , y = 'total_bill' ,color = 'black')
```
