```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
normal = np.random.rand(12,13)
normal
```

```python
sns.heatmap(normal)
```

```python
sns.heatmap(normal , annot = True)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(normal , annot = True , vmin = 0, vmax = 4)
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.heatmap(tips.corr() , annot = True)
```

```python
sns.heatmap(tips.corr() , annot = True , linewidth = 0.7 , vmin = 0 , vmax = 2)
```

```python
sns.heatmap(tips.corr() , annot = True , linewidth = 0.7 , vmin = 0 , vmax = 2 , cmap = 'summer' )
```
