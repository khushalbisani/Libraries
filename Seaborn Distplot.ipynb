```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
num = np.random.randn(150)
```

```python
sns.distplot(num)
```

```python
sns.distplot(num , color = 'violet')
```

```python
# for labeling data
label_dist = pd.Series(num , name = 'variables x')
```

```python
sns.distplot(label_dist)
```

```python
sns.distplot(label_dist , vertical = True)
```

```python
sns.distplot(label_dist , hist = False)
```

```python
sns.distplot(label_dist , rug = True , hist = False)
```
