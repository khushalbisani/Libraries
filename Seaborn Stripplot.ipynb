```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.stripplot(data = tips , x = 'tip')
```

```python
sns.stripplot(data = tips , x = 'day' , y = 'total_bill', palette = 'autumn')
```

```python
sns.stripplot(data = tips , x = 'day' , y = 'total_bill', palette = 'autumn' , jitter = True)
```

```python
sns.stripplot(data = tips , x = 'day' , y = 'total_bill', palette = 'autumn' , jitter = 0.3)
```

```python
sns.stripplot(data = tips , x = 'day' , y = 'total_bill', palette = 'autumn' , jitter = 0.3 , linewidth = 2)
```

```python
sns.stripplot(data = tips , x = 'day' , y = 'total_bill',hue = 'sex', palette = 'autumn' , jitter = 0.1)
```

```python
sns.stripplot(data = tips , x = 'day' , y = 'total_bill',hue = 'sex', palette = 'spring' , jitter = 0.1 ,split = True)
```

```python
sns.stripplot(data = tips , x = 'day' , y = 'total_bill' , hue = 'sex' ,marker = 'd' ,size = 10 ,split = True)
```

```python
sns.stripplot(data = tips , x = 'day' , y = 'total_bill' , hue = 'sex' ,marker = 'd' ,size = 15 ,split = True ,\
              edgecolor = 'grey',alpha = 0.5 ,color = 'blue')
```

```python
sns.stripplot(data = tips ,x ='day', y = 'total_bill',jitter  =True)
sns.boxplot(data = tips ,x ='day', y = 'total_bill',saturation =0.5)
```

```python
sns.stripplot(data = tips ,x ='day', y = 'total_bill',jitter  =True)
sns.violinplot(data = tips ,x ='day', y = 'total_bill',saturation =0.5)
```

```python
sns.stripplot(data = tips ,x ='day', y = 'total_bill',jitter  =True)
sns.violinplot(data = tips ,x ='day', y = 'total_bill',color = '0.9')
```
