```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
mean = [0,0]
cov = [[0.2,0] , [0,3]]
```

```python
x_axis , y_axis = np.random.multivariate_normal(mean,cov,size=5).T
```

```python
sns.kdeplot(x_axis , color = 'red' , shade = True)
```

```python
sns.kdeplot(y_axis , color = 'blue' , shade = True)
```

```python
sns.kdeplot(x_axis , y_axis)
```

```python
sns.kdeplot(x_axis , y_axis , cmap='rocket' , n_levels=20)
```

```python
sns.kdeplot(x_axis , bw = 0.8)
```

```python
sns.kdeplot(x_axis , bw = 0.3)
```

```python
sns.kdeplot(x_axis , bw = 0.1)
```

```python
sns.kdeplot(x_axis , bw = 0.1 , vertical = True)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.species.unique()
```

```python
setosa = iris.loc[iris.species == 'setosa']
setosa.head()
```

```python
virginica = iris.loc[iris.species == 'virginica']
virginica.head()
```

```python
versicolor = iris.loc[iris.species == 'versicolor']
versicolor.head()
```

```python
x = sns.kdeplot( setosa.petal_length , setosa.petal_width ,cmap = 'rocket')
x = sns.kdeplot( versicolor.petal_length , versicolor.petal_width ,cmap = 'autumn'  )
x = sns.kdeplot( virginica.petal_length , virginica.petal_width ,cmap = 'spring' )
```
