```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['Bosch','Amazon','BON','Qubole']
revenue=[34,124,56,76]
profit=[15,36,35,12]
```

```python
new=np.arange(len(company))
new
```

```python
plt.bar(new,revenue,label='revenue')
plt.legend()
```

```python
plt.bar(new,revenue,label='revenue')
plt.xticks(new,company)
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.legend()
```

```python
plt.bar(new,revenue,label='revenue')      #bar = Vertical Bar
plt.bar(new,profit,label='profit')
plt.xticks(new,company)
plt.xlabel('Company')
plt.ylabel('Revenue & Profit')
plt.legend()
```

```python
plt.bar(new+0.2,revenue,width=0.4,label='revenue')
plt.bar(new-0.2,profit,width=0.4,label='profit')
plt.xticks(new,company)
plt.xlabel('Company')
plt.ylabel('Revenue & Profit')
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.barh(new,revenue,label='revenue')  # barh = bar horizontal
plt.barh(new,profit,label='profit')
plt.yticks(new,company)   #yticks = aligned at y axis
plt.legend()
```

```python
plt.barh(new+0.2,revenue,height=0.4,label='revenue')  # barh = bar horizontal
plt.barh(new-0.2,profit,height=0.4,label='profit')
plt.yticks(new,company)   #yticks = aligned at y axis
plt.legend()
```
