# Arithmetic operation

```python
def main():
    subpath = "C:\\Users\\khush\\Documents\\OpenCV\\misc\\"
    
    imgpath1 = subpath + "4.1.03.tiff"
    imgpath2 = subpath + "4.1.04.tiff"
    
    img1 = cv2.imread(imgpath1)
    img2 = cv2.imread(imgpath2)
    
    img1 = cv2.cvtColor(img1 , cv2.COLOR_BGR2RGB)
    img2 = cv2.cvtColor(img2 , cv2.COLOR_BGR2RGB)
    
    add = img1 + img2
    sub1 = img1 - img2
    sub2 = img2 - img1
    mult = img1 * img2
    div1 = img1/img2
    div2 = img2/img1
    
    images = [img1 , img2 , add , sub1 , sub2 , mult , div1 , div2]
    title = ["IMG1" , "IMG2" , "ADD" , "SUB1" , "SUB2" , "MULT" , "DIV1" , "DIV2"]
    
    plt.figure(figsize = (20,20))
    for i in range(8):
        plt.subplot(8,1,i+1)
        plt.imshow(images[i])
        plt.xticks([])
        plt.yticks([])
        plt.title(title[i])
```

```python
if __name__ == "__main__":
    main()
```

# Splitting and Merging Color Image of channel

```python
def main():
    subpath = "C:\\Users\\khush\\Documents\\OpenCV\\misc\\"
    
    imgpath = subpath + "4.1.08.tiff"
    
    img = cv2.imread(imgpath)
    
    img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
    
    r , g , b = cv2.split(img)
    
    images = [cv2.merge((r,g,b)) , r , g , b ]
    title = ["Original" , "Red" , "Green" , "Blue"]
    c_map = ['gray' , 'Reds' , 'Greens' , 'Blues']
    
    plt.figure(figsize=(10,10))
    for i in range(4):
        plt.subplot(4,1,i+1)
        plt.imshow(images[i] , cmap = c_map[i])
        plt.title(title[i])
        plt.xticks([])
        plt.yticks([])
    
```

# Shifting and Translation  || Rotation operation

```python
def main():
    subpath = "C:\\Users\\khush\\Documents\\OpenCV\\misc\\"
    imgpath = subpath + "house.tiff"
    img = cv2.imread(imgpath)
    img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
    
    row , column , channel = img.shape
    
    # Shifting and translation
    T = np.float32([[1,0,-50],[0,1,-50]])
    img1 = cv2.warpAffine(img , T , (column , row))
    
    #In rotation parameter -> (center coordinates , angle , scaling)
    R = cv2.getRotationMatrix2D((column/2 , row/2) ,70, 0.7)
    img2 = cv2.warpAffine(img , R , (column , row))
    
    image = [ img , img1 ,img2]
    title = ["Normal" , "Shifted" , "Rotation"]
    
    plt.figure(figsize=(20,20))
    for i in range(3):
        plt.subplot(3,1,i+1)
        plt.imshow(image[i])
        plt.title(title[i])
```

```python
if __name__ == "__main__":
    main()
```

# MINIPROJECT - Rotation on live web cam feed

```python
import time
def main():
    subpath = "C:\\Users\\khush\\Documents\\OpenCV\\misc\\"
    imgpath = subpath + "house.tiff"
    img = cv2.imread(imgpath)
    img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
    
    row , column , channel = img.shape
    
    angle = 0
    while True:
        if angle == 360:
            angle=0
            
        R = cv2.getRotationMatrix2D((column/2 , row/2) , angle , 0.7)
        
        img1 = cv2.warpAffine(img , R , (column , row))
        
        cv2.imshow("Live Rotation" , img1)
        angle = angle + 1
        time.sleep(0.02)
        if cv2.waitKey(1)==27:
            break
            
    cv2.destroyAllWindows()
```

```python
if __name__ == "__main__":
    main()
```
