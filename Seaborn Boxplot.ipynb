```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.describe()
```

```python
sns.boxplot(data = tips , x='size')
```

```python
sns.boxplot(data = tips , x='total_bill' )
```

```python
sns.boxplot(data = tips , x='sex' , y='total_bill' )
```

```python
#plt.figure(figsize=(8,8))
sns.boxplot(data = tips , x='sex',y='tip')
```

```python
sns.boxplot(data = tips , x='day' , y='total_bill' , hue='sex' , palette = 'spring')
```

```python
sns.boxplot(data = tips , x='day' , y='total_bill' , hue='sex' , color = 'blue')
```

```python
sns.boxplot(data = tips , x='day' ,y='total_bill' ,hue='smoker' )
```

```python
sns.boxplot(data = tips , x='day' ,y='total_bill' ,hue='time' )
```

```python
sns.boxplot(data = tips , x='day' , y='tip' , order = ['Sun','Sat','Fri','Thur'])
```

```python
sns.boxplot(data = tips , x='sex' , y='total_bill' , order = ['Female','Male'])
```
