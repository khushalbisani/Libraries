```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

```python
sns.pairplot(data=iris)
```

```python
sns.pairplot(data=tips)
```

```python
#color , marker ,  vars , hue , palette
sns.pairplot(data=iris , markers = ['+' , 'D' , 's'] , hue = 'species')
```

```python
sns.pairplot(data=iris , markers = ['+' ,'D', 's'] , vars = ['sepal_length' , 'petal_length'] , hue = 'species')
```

```python
sns.pairplot(data=iris , markers = ['+' ,'D', 's'] , vars = ['sepal_width' , 'petal_width'] , hue = 'species', palette = 'autumn')
```

```python
sns.pairplot(data=iris , hue = 'species', palette = 'rocket', \
             x_vars = ['sepal_width' , 'petal_width'] ,\
             y_vars = ['sepal_length' , 'petal_length'])
```

```python
sns.pairplot(data = iris , diag_kind = 'kde')
```

```python
sns.pairplot(data = iris , diag_kind = 'kde' , palette = 'spring' , hue = 'species')
```

```python
sns.pairplot(data = iris , kind = 'reg' )
```

```python
sns.pairplot(data = iris , diag_kind = 'kde' , palette = 'spring' , hue = 'species' , kind = 'reg')
```
